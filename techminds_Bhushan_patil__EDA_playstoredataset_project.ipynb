**Columns which require conversion:**

1) Reviews should be numeric(integer) instead of object

2) Installs should not have "+" and "," signs and must be integer

3) Price should also be converted from object dtype to float and remove '$' sign

4) Last_Updated should be a date instead of Object

```python
#Lets see install column as it is of object data type lets first it convert to int or float
df1['Installs'] = df1.Installs.str.replace(",","")  #using replace function to replace the character in column
df1['Installs'] = df1.Installs.str.replace("+","")
df1['Installs'] = df1.Installs.replace("Free",np.nan)
df1['Installs'] = df1['Installs'].astype(float)
df1['Installs'].dtype
```

```python
# converting price value from object to float so it can be easy to analys
df1['Price'] = df1.Price.replace("Everyone",np.nan)     #using replace function to replace the character in column
df1['Price'] =  df1.Price.str.replace("$","").astype(float)
df1['Price'].dtype    #shows dtype of column
```

```python
# converting size column from object to int
df1['Size'] = df1.Size.replace("Varies with device",np.nan)  #using replace function to replace the character in column
df1['Size'] = df1.Size.str.replace("M","000") # All size values became the kilobyte type.
df1['Size'] = df1.Size.str.replace("k","")
df1['Size'] = df1.Size.replace("1,000+",1000)
df1['Size'] = df1['Size'].astype(float)
df1['Size'].dtype   #shows dtype of column
```

```python
#converting Reviews value from object to float so it can be easy to analys
df1['Reviews'] = df1.Reviews.replace("0.0",0)
df1['Reviews'] = df1.Reviews.replace("3.0M",3000000.0)
df1['Reviews'] = df1['Reviews'].astype(float)
df1['Reviews'].dtype  #shows dtype of column
```

### **checking distribution of  numerical data**



```python
# list of numerical variables
numerical_features = [feature for feature in df1.columns if ((df1[feature].dtypes != 'O') & (feature not in ['y']))]
print('Number of numerical variables: ', len(numerical_features))

# visualise the numerical variables
df1[numerical_features].head()
```

```python
#plot a univariate distribution of continues observations
plt.figure(figsize=(20,60), facecolor='white')
plotnumber =1
for i in numerical_features:         #using for loop to plot the grap
    ax = plt.subplot(12,3,plotnumber)
    sns.distplot(df1[i])          #ploting distribution plot to see the distribution
    plt.xlabel(i)
    plotnumber+=1
plt.show()
```

```python
#using box plot to see the distribution of data
plt.figure(figsize=(20,20))    #setting fig size
plt.subplot(3,3,1)
sns.boxplot(x='Rating',data=df1)   #ploting box plot using seaborn
plt.title("plot no 1")
plt.subplot(3,3,2)
sns.boxplot(x='Reviews',data=df1)
plt.title("plot no 2")
plt.subplot(3,3,3)
sns.boxplot(x='Size',data=df1)
plt.title("plot no 3")
plt.subplot(3,3,4)
sns.boxplot(x='Installs',data=df1)
plt.title("plot no 4")
plt.subplot(3,3,5)
sns.boxplot(x='Price',data=df1)
plt.title("plot no 5")
```

```python
#filling null values for numeric feature
for i in numerical_features:                         #using for loop
  df1[i].fillna(df1[i].median(), inplace=True)       #using fillna function to fill null values
df1
```

```python
#filling null values for v=catagorical data
df1["Category"].fillna(df1["Category"].mode()[0], inplace=True)         #replacing null values by mode using fillna function
df1["Type"].fillna(df1["Type"].mode()[0], inplace=True)
df1["Content Rating"].fillna(df1["Content Rating"].mode()[0], inplace=True)
df1["Genres"].fillna(df1["Genres"].mode()[0], inplace=True)
df1["Last Updated"].fillna(df1["Last Updated"].mode()[0], inplace=True)
df1["Current Ver"].fillna(df1["Current Ver"].mode()[0], inplace=True)
df1["Android Ver"].fillna(df1["Android Ver"].mode()[0], inplace=True)
df1
```

#### **1.Find out app share in the playstore category wise?**

```python
# Chart - 1 visualization code
plt.figure(figsize=(18,18))
plt.pie(df1.Category.value_counts(), labels=df1.Category.value_counts().index, autopct='%1.2f%%')
my_circle = plt.Circle( (0,0), 0.50, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.title('% of apps share in each Category', fontsize = 25)
plt.show()
```

#### **2.What is the average Ratings of Apps ?**

```python
# finding Average app ratings using bar plot
colors = ['orange', 'r', 'yellow', 'g', 'm', 'k']
df1['Rating'].value_counts().plot.bar(figsize=(20,8), color = colors )  #ploting bar plot
plt.xlabel('Average rating',fontsize = 15 )             #labeling x axis
plt.ylabel('Number of apps', fontsize = 15)              #labeling y axis
plt.title('Average rating of apps in Playstore', fontsize = 20)  #giving title
plt.legend();
```
